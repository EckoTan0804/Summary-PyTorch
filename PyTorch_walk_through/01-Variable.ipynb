# `Variable`

```
import torch
from torch.autograd import Variable
```

```
tensor = torch.FloatTensor([[1, 2], [3, 4]])
variable = Variable(tensor, requires_grad=True)
```

```
tensor
```

```
variable
```

## `Variable` in autograd

```
t_out = torch.mean(tensor * tensor)
v_out = torch.mean(variable * variable)
```

```
t_out
```

```
v_out
```

```
v_out.backward()
```

```
variable.grad
```
